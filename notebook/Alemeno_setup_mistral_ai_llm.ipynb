```python
from langchain_community.document_loaders import PyPDFLoader
file_path_list = ["/content/goog-10-k-2023 (1).pdf","/content/tsla-20231231-gen.pdf","/content/uber-10-k-2023.pdf"]
list_of_docs = []
for file_path in file_path_list:
    loader = PyPDFLoader(file_path)
    docs = loader.load()
    list_of_docs.append(docs)
```

```python
list_of_docs
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
list_of_splits = []
for doc in list_of_docs:
    splits = text_splitter.split_documents(doc)
    list_of_splits.append(splits)
```

```python
list_of_splits
```

```python
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

model_name = "BAAI/bge-base-en"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}
hf = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)
```

```python
vectorstore_list = []
for splits in list_of_splits:
    vectorstore = Chroma.from_documents(documents=splits, embedding=hf)
    vectorstore_list.append(vectorstore)
```

```python
import getpass
import os

os.environ["MISTRAL_API_KEY"] = getpass.getpass()

from langchain_mistralai import ChatMistralAI

llm = ChatMistralAI(model="mistral-large-latest")
```

```python
def get_context(question):
    context_list = []
    for vectorstore in vectorstore_list:
        docs = vectorstore.similarity_search(question)[-1].page_content
        context_list.append(docs)
    # return context_list
    context = ""
    for i in range(0,len(context_list)):
        # for doc in context_list[i]:
        context += f"doc {i}: " + context_list[i] + "\n\n\n"
    return context
```

```python
from langchain.prompts import ChatPromptTemplate
```

```python
def get_answer(question):
    context = get_context(question)
    message = f"""
              Answer this question using the provided context only. Make sure to answer the questions in more personised way and user friendly way.

              {question}

              Context:
              {context}
              """
    prompt = ChatPromptTemplate.from_messages([("human", message)])
    rag_chain = prompt | llm
    return rag_chain.invoke({}).content
```
