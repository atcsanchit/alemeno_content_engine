```python
output_path = "/content/model"
tokenizer_1 = AutoTokenizer.from_pretrained(output_path)
model_1 = AutoModelForCausalLM.from_pretrained(output_path)
```

```python
prompt = """Instruction: you are a helpful assistant.
Question: Can you tell me what is python programming in 4 sentences?
Answer:
"""

# Tokenize the input prompt
input_ids = tokenizer_1(prompt, return_tensors="pt").input_ids

# Generate a response
outputs = model_1.generate(input_ids, max_length=50, num_beams=5, early_stopping=True)

# Decode the generated tokens to text
response = tokenizer.decode(outputs[0], skip_special_tokens=True)

# Print the response
print("Generated Response:", response)
```
