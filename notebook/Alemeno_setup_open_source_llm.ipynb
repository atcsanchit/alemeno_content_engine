```python
from transformers import AutoTokenizer, AutoModelForCausalLM
# from transformers import Trainer, TrainingArguments
```

```python
model_name = "Qwen/Qwen2.5-0.5B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
output_path = "/content/model"

model.save_pretrained(output_path)
tokenizer.save_pretrained(output_path)
```

```python
output_path = "/content/model"
tokenizer_1 = AutoTokenizer.from_pretrained(output_path)
model_1 = AutoModelForCausalLM.from_pretrained(output_path)
```

```python
prompt = """Instruction: you are a helpful assistant. Make sure you always answer in Pirate language.
Question: Can you tell me what is python programming in 4 sentences?
Answer:
"""

input_ids = tokenizer_1(prompt, return_tensors="pt").input_ids

outputs = model_1.generate(input_ids, max_length=50, num_beams=5, early_stopping=True)

response = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Generated Response:", response)
```
